```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('novels')
df2 = pd.read_csv('novel_rating')
```

```python
df2.head()
```

```python
data2 = df2.drop('Unnamed: 0',axis =1)
```

```python
df1["Published Year"].unique()
```

```python
df1['Published Year'] = df1['Published Year'].apply(lambda x: str(x).replace('\r',''))
```

```python
data = df1.drop('Unnamed: 0',axis =1)
```

```python
from sklearn.preprocessing import LabelEncoder

combined_titles = list(data['Title']) + list(data2['Title'])
encoder = LabelEncoder()
encoder.fit(combined_titles)
data["Title id"] = encoder.transform(data['Title'])
data2["Title id"] = encoder.transform(data2['Title'])
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

data2["User id"] = encoder.fit_transform(data2['User Name'])
```

```python
data2.head()
```

```python
user_id_counts = data2['User id'].value_counts()

print(user_id_counts)
```

```python
data.to_csv("novels_clean")
data2.to_csv("novel_rating_clean")
```
